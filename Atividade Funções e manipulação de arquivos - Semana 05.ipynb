```python
def verificar_paridade(numero):
    return numero % 2 == 0

#Exemplos:
print(verificar_paridade(85))
print(verificar_paridade(200))
```

Escreva uma função chamada verificar_palindromo que recebe uma string como parâmetro e retorna True se for um palíndromo (uma palavra que é igual se lida da esquerda para a direita e vice-versa) e False caso contrário.

```python
def verificar_palindromo(string):
    
    #Verificando palíndromo:
    return string == string[::-1]

#Exemplo:
print(verificar_palindromo("ovo"))
```

```python
def verificar_palindromo(string):

    #Removendo os espaços em branco e tornando todas as letras minúsculas para garantir uma comparação correta:
    string = string.replace(" ", "").lower()
    
    return string == string[::-1]

#Exemplos:
print(verificar_palindromo("arara"))
print(verificar_palindromo("python"))
```

##Desafio listas e funções:

Crie uma função que representa o jogo batalha naval.

Dentro da função crie uma estrutura bidimensional utilizando listas com sublistas para representar um tabuleiro (1 lista com 20 elementos e cada elemento é uma lista de 20 elementos, tabuleiro 20x20). Cada posição irá armazenar 1 valor numérico que significa:
- 0 - Água
- 1 - Navio

Para cada posição escolha esses valores aleatoriamente, respeitando a regra de que não podem existir mais de 20 navios no tabuleiro.

Após os valores serem distribuídos, o programa deve pedir ao usuário uma posição do tabuleiro e informar se ele acertou um navio ou água e repetir o procedimento até que o usuário derrote todos os navios ou chegue ao limite de 35 tentativas.



```python
import random

# Criação do tabuleiro
tabuleiro = [[0 for _ in range(20)] for _ in range(20)]
navios_colocados = 0

print("Tabuleiro:")
for row in tabuleiro:
    print(row)



# Colocação dos navios no tabuleiro
while navios_colocados < 20:
    x = random.randint(0, 19)
    y = random.randint(0, 19)
    if tabuleiro[x][y] == 0:
        tabuleiro[x][y] = 1
        navios_colocados += 1

print("Tabuleiro com navios:")
for row in tabuleiro:
    print(row)

# Jogo
tentativas = 0
while tentativas < 35:
    print("Tentativa", tentativas + 1)

    x = int(input("Digite a coordenada X (1-20): "))-1
    while x <0 or x > 19:
        print("Coordenadas inválidas. Tente novamente.")
        x = int(input("Digite a coordenada X (1-20): "))-1

    y = int(input("Digite a coordenada Y (1-20): "))-1
    while y <0 or y > 19:
        print("Coordenadas inválidas. Tente novamente.")
        y = int(input("Digite a coordenada X (1-20): "))-1

#    print(x,y)


    if tabuleiro[x][y] == 1:
        print("Você acertou um navio!")
        tabuleiro[x][y] = 0
        if sum(sum(row) for row in tabuleiro) == 0:
            print("Parabéns, você afundou todos os navios!")
            break
    else:
        print("Você errou!")

    tentativas += 1
else:
    print("Você excedeu o limite de tentativas. Game over.")
```

```python
import random

# Criação do tabuleiro
tabuleiro = [[0 for _ in range(20)] for _ in range(20)]
navios_colocados = 0

print("Tabuleiro:")
for row in tabuleiro:
    print(row)

# Colocação dos navios no tabuleiro
while navios_colocados < 20:
    x = random.randint(0, 19)
    y = random.randint(0, 19)
    if tabuleiro[x][y] == 0:
        tabuleiro[x][y] = 1
        navios_colocados += 1

print("Tabuleiro com navios:")
for row in tabuleiro:
    print(row)

# Jogo
tentativas = 0
acertos_agua = 0
acertos_navios = 0
max_acertos_ininterruptos = 0
acertos_ininterruptos = 0

while tentativas < 35:
    print("Tentativa", tentativas + 1)

    x = int(input("Digite a coordenada X (1-20): ")) - 1
    while x < 0 or x > 19:
        print("Coordenadas inválidas. Tente novamente.")
        x = int(input("Digite a coordenada X (1-20): ")) - 1

    y = int(input("Digite a coordenada Y (1-20): ")) - 1
    while y < 0 or y > 19:
        print("Coordenadas inválidas. Tente novamente.")
        y = int(input("Digite a coordenada X (1-20): ")) - 1

    if tabuleiro[x][y] == 1:
        print("Você acertou um navio!")
        acertos_navios += 1
        acertos_ininterruptos += 1
        max_acertos_ininterruptos = max(max_acertos_ininterruptos, acertos_ininterruptos)
        tabuleiro[x][y] = 0
        if sum(sum(row) for row in tabuleiro) == 0:
            print("Parabéns, você afundou todos os navios!")
            break
    else:
        print("Você acertou água!") #Você errou!
        acertos_agua += 1
        acertos_ininterruptos = 0

    tentativas += 1

else:
    print("Você excedeu o limite de tentativas. Game over.")

# Calculando as estatísticas:
total_acertos = acertos_agua + acertos_navios
porcentagem_acertos_agua = (acertos_agua / total_acertos) * 100 if total_acertos > 0 else 0
porcentagem_acertos_navios = (acertos_navios / total_acertos) * 100 if total_acertos > 0 else 0

print("\nEstatísticas:")
print("Acertos em água:", acertos_agua)
print("Acertos em navios:", acertos_navios)
print("Porcentagem de acertos em água: {:.2f}%".format(porcentagem_acertos_agua))
print("Porcentagem de acertos em navios: {:.2f}%".format(porcentagem_acertos_navios))
print("Acertos ininterruptos máximos:", max_acertos_ininterruptos)
```

Procure quem foi a pessoa que mais gastou?

```python
import pandas as pd

arquivo_csv = 'C:\\Users\\Milene\\Documents\\Engenharia de Dados\\Introdução a Programação - Python 1\\Atividade Funções\\compras.csv'

df = pd.read_csv(arquivo_csv)

# Inspeção do DataFrame para verificar as colunas
print(df.head())

# Definição das faixas etárias
def faixa_etaria(idade):
    if 18 <= idade <= 25:
        return 'Jovens'
    elif 26 <= idade <= 59:
        return 'Adultos'
    elif idade >= 60:
        return 'Idosos'
    else:
        return 'Fora da faixa etária'

# Função para classificar as faixas etárias
df['faixa_etaria'] = df['idade'].apply(faixa_etaria)

# Concatenando nome e sobrenome para criar uma coluna 'nome_completo'
df['nome_completo'] = df['nome'] + ' ' + df['sobrenome']

# Cálculo o gasto total por pessoa
gastos_totais = df.groupby(['nome_completo', 'faixa_etaria'])['compra'].sum().reset_index()

maiores_gastos = gastos_totais.loc[gastos_totais.groupby('faixa_etaria')['compra'].idxmax()]

print(maiores_gastos)

```

Busque quais são os anos da base de dados?

```python
anos_unicos = df['ano'].unique()

# Ordenando os anos
anos_unicos_ordenados = sorted(anos_unicos)

print("Anos presentes na base de dados:", anos_unicos_ordenados)
```

Qual a porcentagem de homens e mulheres na base de dados?

```python
contagem_sexo = df['sexo'].value_counts()

# Cálculo da porcentagem de homens e mulheres
total_registros = len(df)
porcentagem_homens = (contagem_sexo['M'] / total_registros) * 100
porcentagem_mulheres = (contagem_sexo['F'] / total_registros) * 100

print(f"Porcentagem de homens na base de dados: {porcentagem_homens:.2f}%")
print(f"Porcentagem de mulheres na base de dados: {porcentagem_mulheres:.2f}%")
```

Qual foi o gasto por ano?

```python
# Dados agrupados por ano e soma dos valores das compras para cada ano
gastos_por_ano = df.groupby('ano')['compra'].sum()

print("Gasto por ano:")
print(gastos_por_ano)
```

Qual foi o ano com maior gasto?

```python
gastos_por_ano = df.groupby('ano')['compra'].sum()

# Ano com o maior gasto
ano_maior_gasto = gastos_por_ano.idxmax()
maior_gasto = gastos_por_ano.max()

print(f"Ano com maior gasto: {ano_maior_gasto}")
print(f"Gasto total nesse ano: {maior_gasto}")
```

Utilizando as faixas etárias, diga quantas pessoas há em cada faixa?

```python
# Função para classificar as faixas etárias
df['faixa_etaria'] = df['idade'].apply(faixa_etaria)

# Número de pessoas em cada faixa etária
contagem_faixa_etaria = df['faixa_etaria'].value_counts()

print("Contagem de pessoas em cada faixa etária:")
print(contagem_faixa_etaria)
```

Qual é a faixa etária que mais gasta?

```python
# Cálculo do gasto total por faixa etária
gastos_por_faixa_etaria = df.groupby('faixa_etaria')['compra'].sum()

# Faixa etária que mais gasta
faixa_etaria_mais_gasta = gastos_por_faixa_etaria.idxmax()
maior_gasto = gastos_por_faixa_etaria.max()

print(f"Faixa etária que mais gasta: {faixa_etaria_mais_gasta}")
```

A opção de débito é mais utilizada entre homens ou mulheres?

```python
# Número de transações de débito para homens e mulheres
contagem_debito_homens = df.loc[(df['pagamento'] == 'debito') & (df['sexo'] == 'M')].shape[0]
contagem_debito_mulheres = df.loc[(df['pagamento'] == 'debito') & (df['sexo'] == 'F')].shape[0]

# Sexo que utilizou mais a função de débito
if contagem_debito_homens > contagem_debito_mulheres:
    print("A função de débito é mais utilizada entre homens.")
elif contagem_debito_homens < contagem_debito_mulheres:
    print("A função de débito é mais utilizada entre mulheres.")
else:
    print("Homens e mulheres utilizam igualmente a função de débito.")
```

Qual o sobrenome que mais aparece na base de dados?

```python
sobrenome_mais_comum = df['sobrenome'].value_counts().idxmax()

print("O sobrenome mais comum na base de dados é:", sobrenome_mais_comum)
```

Qual foi o ano em que os homens mais usaram o crédito?

```python
# Transações feitas por homens com pagamento por crédito
transacoes_credito_homens = df[(df['sexo'] == 'M') & (df['pagamento'] == 'credito')]

# Cálculo do total de transações de crédito para cada ano
total_transacoes_credito_por_ano = transacoes_credito_homens.groupby('ano').size()

# Ano com o maior número de transações de crédito para homens
ano_mais_transacoes_credito_homens = total_transacoes_credito_por_ano.idxmax()
total_transacoes_credito_mais_homens = total_transacoes_credito_por_ano.max()

print(f"O ano em que os homens mais usaram crédito foi {ano_mais_transacoes_credito_homens}, com um total de {total_transacoes_credito_mais_homens} utilizações.")
```

Qual opção de pagamento é mais utilizada em cada faixa etária?

```python
# Função para classificar as faixas etárias
df['faixa_etaria'] = df['idade'].apply(faixa_etaria)

# Dados agrupados pela combinação de faixa etária e forma de pagamento e contagem do número de ocorrências em cada grupo
contagem_forma_pagamento_por_faixa_etaria = df.groupby(['faixa_etaria', 'pagamento']).size()

# Forma de pagamento mais utilizada em cada faixa etária
forma_pagamento_mais_utilizada_por_faixa_etaria = contagem_forma_pagamento_por_faixa_etaria.groupby(level=0).idxmax().reset_index()

print("Forma de pagamento mais utilizada em cada faixa etária:")
print(forma_pagamento_mais_utilizada_por_faixa_etaria)
```

Qual o valor gasto em compras por jovens do ano de 2010 até 2015?

```python
# Filtro das transações feitas por jovens nos anos de 2010 a 2015
transacoes_jovens_2010_2015 = df[(df['faixa_etaria'] == 'Jovens') & (df['ano'] >= 2010) & (df['ano'] <= 2015)]

# Cálculo do valor total gasto em compras por jovens nos anos de 2010 a 2015
valor_total_gasto_jovens_2010_2015 = transacoes_jovens_2010_2015['compra'].sum()

print("Valor total gasto em compras por jovens nos anos de 2010 a 2015: R$", valor_total_gasto_jovens_2010_2015)
```
